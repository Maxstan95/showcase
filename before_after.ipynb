```python
import pandas as pd
from scipy import stats
import json
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('default')
mpl.rcParams['agg.path.chunksize'] = 10000
```

```python
with open('../exp_pairs.json', 'r') as fp:
    exp_pairs = json.load(fp)
```

```python
with open('../contr_pairs.json', 'r') as fp:
    contr_pairs = json.load(fp)
```

```python
for i in exp_pairs:

    stat, p = stats.ttest_rel(exp_pairs[i]['iter1'], exp_pairs[i]['iter2'])
    if p<0.05:
        print('T-test for repeated samples: p-val < 0.05')
    else:
        print('T-test for repeated samples: p-val > 0.05')
    
    data_to_plot = [exp_pairs[i]['iter1'], exp_pairs[i]['iter2']]
    fig = plt.figure(1, figsize=(6, 5))
    ax = fig.add_subplot(111)
    ax.set_title('Эксп. '+i)
    ax.grid(True, axis='y')
    
    bp = ax.boxplot(data_to_plot, notch=False, showmeans=True, meanline=True)
    ax.set_xticklabels(['До курса','После курса'])
    ax.set_ylabel('Балл')
    plt.show()
```

```python
for i in contr_pairs:

    stat, p = stats.ttest_rel(contr_pairs[i]['iter1'], contr_pairs[i]['iter2'])
    if p<0.05:
        print('T-test for repeated samples: p-val < 0.05')
    else:
        print('T-test for repeated samples: p-val > 0.05')
    
    data_to_plot = [contr_pairs[i]['iter1'], contr_pairs[i]['iter2']]
    fig = plt.figure(1, figsize=(6, 5))
    ax = fig.add_subplot(111)
    ax.set_title('Контрольн. '+i)
    ax.grid(True, axis='y')
    
    bp = ax.boxplot(data_to_plot, notch=False, showmeans=True, meanline=True)
    ax.set_xticklabels(['Весна','Начало лета'])
    ax.set_ylabel('Балл')
    plt.show()
```
