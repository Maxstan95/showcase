```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('default')
mpl.rcParams['agg.path.chunksize'] = 10000
```

```python
contr_delta_df = pd.read_csv('../contr.csv')
```

```python
exp_delta_df = pd.read_csv('../exp.csv')
```

```python
for i in exp_delta_df.columns[4:]:
    
    data_to_plot = [exp_delta_df[i], contr_delta_df[i]]
    fig = plt.figure(1, figsize=(8, 6))
    ax = fig.add_subplot(111)
    ax.set_title(" "+ i)
    ax.grid(True, axis='y')
    bp = ax.boxplot(data_to_plot, notch=True)
    ax.set_xticklabels(['',''])
    fig.tight_layout()
    ax.set_ylabel(' ')
    plt.show()
```
